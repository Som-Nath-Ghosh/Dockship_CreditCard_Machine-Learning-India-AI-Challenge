```python
outliers=[]
def det_outliers(data):
    threshold =3
    mean=np.mean(data)
    stdev=np.std(data)
    
    for i in data:
        z_score=(i-mean)/stdev
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```
